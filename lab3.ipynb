```python
# ===== IMPORTS & CONFIG =====
import os, sys, math, time, json
from collections import Counter, defaultdict
import pandas as pd
from tqdm import tqdm

# radon
from radon.raw import analyze as radon_raw_analyze
from radon.complexity import cc_visit
from radon.metrics import mi_visit

# embeddings & similarity
import torch
from transformers import AutoTokenizer, AutoModel

# BLEU
import sacrebleu

# numeric / ml utils
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# ===== SETTINGS =====
INPUT_FILENAME = None  # set to None to auto-discover common names
CHECKPOINT_EVERY = 100  # save intermediate CSV every N rows (change as needed)
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# ===== Auto-detect input CSV in Kaggle environment =====
candidates = [
    "diff_analysis_with_llm.csv",
    "diff_analysis_with_llm_metrics.csv",
    "diffs_final.csv",
    "diff_analysis_with_llm_full_metrics.csv",
    "diffs_with_llm.csv",
    "diff_analysis_with_llm_classified.csv"
]
# Also check /kaggle/input subfolders
kaggle_input = "/kaggle/input"
if os.path.isdir(kaggle_input):
    for root, dirs, files in os.walk(kaggle_input):
        for f in files:
            if f.lower().endswith(".csv"):
                candidates.append(os.path.join(root, f))

if INPUT_FILENAME is None:
    found = None
    for c in candidates:
        if os.path.exists(c):
            found = c
            break
    if found is None:
        raise FileNotFoundError("No input CSV found. Upload your Lab2 dataset to Kaggle working directory and name it e.g. diff_analysis_with_llm.csv")
    INPUT_FILENAME = found

print("Using input:", INPUT_FILENAME)

# ===== Load dataset =====
df = pd.read_csv(INPUT_FILENAME, encoding="utf-8")
print("Loaded rows:", len(df))
# normalize column names (common variants)
colmap = {c: c.strip() for c in df.columns}
df.rename(columns=colmap, inplace=True)

# helper to find column names robustly
def find_col(df, candidates):
    cols = df.columns
    for cand in candidates:
        if cand in cols:
            return cand
    # lower-case match
    lc = {c.lower(): c for c in cols}
    for cand in candidates:
        if cand.lower() in lc:
            return lc[cand.lower()]
    return None

# probable column names (adjust if different)
HASH_COL = find_col(df, ["Hash", "hash", "CommitHash"])
MSG_COL  = find_col(df, ["Message", "Msg", "Commit Message", "commit_message"])
FNAME_COL = find_col(df, ["Filename", "File", "new_path", "ModifiedFile"])
SRC_BEFORE_COL = find_col(df, ["Source Code (before)", "Source Code before", "SourceCode_Before", "Source_before", "source_before"])
SRC_AFTER_COL  = find_col(df, ["Source Code (current)", "Source Code (after)", "Source Code current", "SourceCode_After", "Source_after", "source_after"])
LLM_FIX_COL = find_col(df, ["LLM Inference (fix type)", "LLM Inference", "LLM_Inference", "LLM Inference (fix type)"])

print("Columns used:", HASH_COL, MSG_COL, FNAME_COL, SRC_BEFORE_COL, SRC_AFTER_COL, LLM_FIX_COL)

# If a column is missing, stop early so user can adjust
needed = [HASH_COL, MSG_COL, FNAME_COL, SRC_BEFORE_COL, SRC_AFTER_COL]
if any(x is None for x in needed):
    missing = [name for name, val in zip(["Hash","Message","Filename","Source(before)","Source(after)"], needed) if val is None]
    raise ValueError(f"Required columns missing: {missing}. Rename your CSV columns to include these fields.")

# ===== 2A. Baseline descriptive statistics =====
total_rows = len(df)
unique_commits = df[HASH_COL].nunique()
total_files = total_rows
files_per_commit = df.groupby(HASH_COL).size()
avg_files_per_commit = files_per_commit.mean()
top_extensions = df[FNAME_COL].dropna().apply(lambda x: os.path.splitext(str(x))[-1].lower() or "(noext)").value_counts().head(10)
fix_type_dist = None
if LLM_FIX_COL:
    fix_type_dist = df[LLM_FIX_COL].fillna("UNKNOWN").value_counts()

print("Baseline stats:")
print("Total rows (file-level):", total_rows)
print("Unique commits:", unique_commits)
print("Total files:", total_files)
print("Avg files per commit: {:.2f}".format(avg_files_per_commit))
print("Top extensions:\n", top_extensions)
if fix_type_dist is not None:
    print("LLM fix-type distribution (top):\n", fix_type_dist.head(10))

# save baseline summary
baseline = {
    "total_rows": total_rows,
    "unique_commits": unique_commits,
    "avg_files_per_commit": float(avg_files_per_commit),
    "top_extensions": top_extensions.to_dict(),
    "fix_type_dist": fix_type_dist.to_dict() if fix_type_dist is not None else {}
}
pd.Series(baseline).to_json("baseline_summary.json")
print("Baseline saved: baseline_summary.json")

# ===== 3) Radon metrics: MI, CC_total (sum of block complexity), LOC (we use sloc) =====
# We'll create/overwrite these columns (keeps dtype flexible)
for c in ["MI_Before","MI_After","CC_Before","CC_After","LOC_Before","LOC_After"]:
    if c not in df.columns:
        df[c] = None

def safe_radon_metrics(code_str):
    """Return dict: mi(float|None), cc_total(int|None), sloc(int|None)"""
    if not isinstance(code_str, str) or code_str.strip() == "":
        return {"mi": None, "cc_total": None, "sloc": None}
    try:
        # raw metrics
        raw = radon_raw_analyze(code_str)
        sloc = int(raw.sloc)  # source lines of code
        # maintainability index
        mi = float(mi_visit(code_str, True))
        # cyclomatic complexity (list of blocks)
        blocks = cc_visit(code_str)
        cc_total = sum(getattr(b, "complexity", 0) for b in blocks) if blocks else 0
        return {"mi": mi, "cc_total": cc_total, "sloc": sloc}
    except Exception as e:
        # catch parse errors, return Nones
        return {"mi": None, "cc_total": None, "sloc": None}

# If metrics were already partially computed in a previous run, we can skip those rows.
start_idx = 0
if os.path.exists("radon_checkpoint.csv"):
    prev = pd.read_csv("radon_checkpoint.csv")
    # align indices by Hash+Filename to resume robustly
    done_keys = set(prev[HASH_COL].astype(str) + "||" + prev[FNAME_COL].astype(str))
    done_mask = df.apply(lambda r: (str(r[HASH_COL]) + "||" + str(r[FNAME_COL])) in done_keys, axis=1)
    start_idx = done_mask.sum()
    print(f"Resuming radon metric computation: {start_idx} rows already done (found radon_checkpoint.csv)")

pbar = tqdm(total=len(df), desc="Radon metrics")
for idx, row in df.iterrows():
    if idx < start_idx:
        pbar.update(1)
        continue
    before_code = row[SRC_BEFORE_COL]
    after_code  = row[SRC_AFTER_COL]
    mb = safe_radon_metrics(before_code)
    ma = safe_radon_metrics(after_code)
    df.at[idx, "MI_Before"] = mb["mi"]
    df.at[idx, "CC_Before"] = mb["cc_total"]
    df.at[idx, "LOC_Before"] = mb["sloc"]
    df.at[idx, "MI_After"] = ma["mi"]
    df.at[idx, "CC_After"] = ma["cc_total"]
    df.at[idx, "LOC_After"] = ma["sloc"]
    # Save periodic checkpoint
    if idx % CHECKPOINT_EVERY == 0 and idx > 0:
        df.to_csv("radon_checkpoint.csv", index=False)
    pbar.update(1)
pbar.close()
# Save final radon-enriched CSV
df.to_csv("diffs_with_radon.csv", index=False, encoding="utf-8")
print("Saved diffs_with_radon.csv")

# Compute change columns
def safe_sub(a,b):
    try:
        return None if (pd.isna(a) or pd.isna(b)) else float(a) - float(b)
    except Exception:
        return None

df["MI_Change"] = df.apply(lambda r: safe_sub(r["MI_After"], r["MI_Before"]), axis=1)
df["CC_Change"] = df.apply(lambda r: safe_sub(r["CC_After"], r["CC_Before"]), axis=1)
df["LOC_Change"] = df.apply(lambda r: safe_sub(r["LOC_After"], r["LOC_Before"]), axis=1)

# quick sanity print
print("Radon metrics computed. Example rows:\n", df[[HASH_COL, FNAME_COL, "MI_Before","MI_After","MI_Change","CC_Before","CC_After","CC_Change","LOC_Before","LOC_After","LOC_Change"]].head())

# ===== 4) Change Magnitude metrics =====
# 4A: Setup CodeBERT (microsoft/codebert-base)
model_name = "microsoft/codebert-base"
print("Loading CodeBERT model:", model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name).to(DEVICE)
model.eval()

def codebert_embed(code: str):
    if not isinstance(code, str) or code.strip() == "":
        return None
    # For long code we truncate; an alternative is chunking + mean pooling across chunks
    inputs = tokenizer(code, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
    with torch.no_grad():
        out = model(**inputs)
    last_hidden = out.last_hidden_state  # (1, seq_len, hidden)
    emb = last_hidden.mean(dim=1).cpu().numpy()  # mean pooling -> shape (1, hidden)
    return emb  # numpy array shape (1, dim)

def semantic_similarity(code_before, code_after):
    emb_b = codebert_embed(code_before)
    emb_a = codebert_embed(code_after)
    if emb_b is None or emb_a is None:
        return None
    return float(cosine_similarity(emb_b, emb_a)[0][0])

# 4B: Token similarity using SacreBLEU (sentence-level)
def token_similarity_sacre(code_before, code_after):
    if not isinstance(code_before, str) or not isinstance(code_after, str):
        return None
    if code_before.strip() == "" or code_after.strip() == "":
        return None
    try:
        # sacrebleu.sentence_bleu returns a Score-like object or float; convert to 0..1
        res = sacrebleu.sentence_bleu(code_after, [code_before])  # candidate, [reference]
        if hasattr(res, "score"):
            return float(res.score) / 100.0
        else:
            # If it returns a float already in 0..100
            val = float(res)
            # Some versions return already 0..1; guard:
            return val/100.0 if val > 1.0 else val
    except Exception as e:
        # fallback: simple token overlap proportion (very rough)
        try:
            a = code_before.split()
            b = code_after.split()
            if len(a) == 0:
                return 0.0
            inter = sum(1 for tok in b if tok in set(a))
            return float(inter) / max(1, len(b))
        except:
            return None

# Add columns for similarities if not present
for c in ["Semantic_Similarity", "Token_Similarity"]:
    if c not in df.columns:
        df[c] = None

# Compute embeddings and BLEU row-by-row with checkpointing
start_idx = 0
if os.path.exists("similarity_checkpoint.csv"):
    prev = pd.read_csv("similarity_checkpoint.csv")
    done_keys = set(prev[HASH_COL].astype(str) + "||" + prev[FNAME_COL].astype(str))
    done_mask = df.apply(lambda r: (str(r[HASH_COL]) + "||" + str(r[FNAME_COL])) in done_keys, axis=1)
    start_idx = done_mask.sum()
    print(f"Resuming similarity computation: {start_idx} rows already done (found similarity_checkpoint.csv)")

pbar = tqdm(total=len(df), desc="Computing similarities")
for idx, row in df.iterrows():
    if idx < start_idx:
        pbar.update(1)
        continue
    cb = row[SRC_BEFORE_COL]
    ca = row[SRC_AFTER_COL]
    try:
        df.at[idx, "Semantic_Similarity"] = semantic_similarity(cb, ca)
    except Exception as e:
        df.at[idx, "Semantic_Similarity"] = None
    try:
        df.at[idx, "Token_Similarity"] = token_similarity_sacre(cb, ca)
    except Exception as e:
        df.at[idx, "Token_Similarity"] = None
    # checkpoint
    if idx % CHECKPOINT_EVERY == 0 and idx > 0:
        df.to_csv("similarity_checkpoint.csv", index=False)
    pbar.update(1)
pbar.close()

df.to_csv("diffs_with_change_magnitude.csv", index=False, encoding="utf-8")
print("Saved diffs_with_change_magnitude.csv")

# ===== 5) Classification & Agreement =====
SEMANTIC_THRESHOLD = 0.80
TOKEN_THRESHOLD = 0.75

def classify_semantic(s):
    if s is None or pd.isna(s):
        return "Unknown"
    return "Minor Fix" if s >= SEMANTIC_THRESHOLD else "Major Fix"

def classify_token(s):
    if s is None or pd.isna(s):
        return "Unknown"
    return "Minor Fix" if s >= TOKEN_THRESHOLD else "Major Fix"

df["Semantic_Class"] = df["Semantic_Similarity"].apply(classify_semantic)
df["Token_Class"] = df["Token_Similarity"].apply(classify_token)
df["Classes_Agree"] = df.apply(lambda r: "YES" if r["Semantic_Class"] == r["Token_Class"] and r["Semantic_Class"] != "Unknown" else ("Unknown" if (r["Semantic_Class"]=="Unknown" or r["Token_Class"]=="Unknown") else "NO"), axis=1)

# Per-commit aggregation example (optionally useful)
# Strategy: if ANY file in commit is "Major Fix" -> commit classified as Major (conservative).
commit_groups = []
for commit, sub in df.groupby(HASH_COL):
    # compute per-commit aggregated similarity as mean (skipping Nones)
    sem_vals = [v for v in sub["Semantic_Similarity"].tolist() if not pd.isna(v)]
    tok_vals = [v for v in sub["Token_Similarity"].tolist() if not pd.isna(v)]
    sem_mean = float(np.mean(sem_vals)) if len(sem_vals)>0 else None
    tok_mean = float(np.mean(tok_vals)) if len(tok_vals)>0 else None
    sem_class = classify_semantic(sem_mean) if sem_mean is not None else "Unknown"
    tok_class = classify_token(tok_mean) if tok_mean is not None else "Unknown"
    agg_agree = "YES" if (sem_class == tok_class and sem_class != "Unknown") else ("Unknown" if sem_class=="Unknown" or tok_class=="Unknown" else "NO")
    commit_groups.append({
        "Hash": commit,
        "Semantic_Mean": sem_mean,
        "Token_Mean": tok_mean,
        "Semantic_Class": sem_class,
        "Token_Class": tok_class,
        "Classes_Agree": agg_agree,
        "Num_Files": len(sub)
    })
commits_df = pd.DataFrame(commit_groups)
commits_df.to_csv("commit_level_classification.csv", index=False)
print("Saved commit-level classification: commit_level_classification.csv")

# Save final file-level CSV
OUTPUT_CSV = "diff_analysis_with_llm_full_metrics.csv"
df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8")
print("Saved final file-level CSV:", OUTPUT_CSV)

# Quick summary stats for agreement
agree_counts = df["Classes_Agree"].value_counts(dropna=False)
print("Agreement counts (file-level):\n", agree_counts)
print("Commit-level agreement top:\n", commits_df["Classes_Agree"].value_counts())

# A couple of helpful plots (requires matplotlib, but Kaggle supports it)
try:
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(1,3, figsize=(15,4))
    df["Semantic_Similarity"].dropna().hist(bins=20, ax=ax[0]); ax[0].set_title("Semantic similarity")
    df["Token_Similarity"].dropna().hist(bins=20, ax=ax[1]); ax[1].set_title("Token similarity")
    df["MI_Change"].dropna().hist(bins=20, ax=ax[2]); ax[2].set_title("MI change (After - Before)")
    plt.tight_layout()
    plt.savefig("quick_histograms.png", dpi=200)
    print("Saved quick_histograms.png")
except Exception as e:
    print("Plotting skipped (matplotlib missing or error):", e)
```
